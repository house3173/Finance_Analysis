```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data with the correct encoding
df = pd.read_excel('LIST_COMPANY.xlsx', header=0)
df.head()
```

```python
## Đưa ra danh sách các mảng chứa mã cổ phiếu nhóm theo ngành nghề (chú ý viết thường hết)

# Tạo mảng chứa mã cổ phiếu nhóm theo ngành nghề
nganh = df["Ngành"].unique()
ma_cp = []
for i in nganh:
    # Chuyển mã cổ phiếu về chữ thường 
    ma_cp.append(df[df["Ngành"] == i]["Mã CP"].str.lower().values)

# In ra danh sách mã cổ phiếu nhóm theo ngành nghề
for i in range(len(nganh)):
    ## Định dạng in có dấu phẩy giữa các mã, mỗi mã được đặt bên trong nháy đơn
    print(nganh[i], ":", ', '.join(["'" + x + "'" for x in ma_cp[i]]))
```

```python
## Đọc dữ liệu từ file excel bctc/Can_doi_ke_toan_VCB.xlsx
df = pd.read_excel('bctc/Can_doi_ke_toan_VCB.xlsx', header=1)
# Hiển thị các header của file excel
print(df.columns)

# Chuyển tất cả các giá trị trong df từ cột 2 trở đi về dạng số, nếu không có giá trị thì chuyển về 0
df.iloc[:, 2:] = df.iloc[:, 2:].apply(pd.to_numeric, errors='coerce').fillna(0)

list_quy = ['Quý 2 - 2018', 'Quý 3 - 2018', 'Quý 4 - 2018', 'Quý 1 - 2019',
    'Quý 2 - 2019', 'Quý 3 - 2019', 'Quý 4 - 2019', 'Quý 1 - 2020',
    'Quý 2 - 2020', 'Quý 3 - 2020', 'Quý 4 - 2020', 'Quý 1 - 2021',
    'Quý 2 - 2021', 'Quý 3 - 2021', 'Quý 4 - 2021', 'Quý 1 - 2022',
    'Quý 2 - 2022', 'Quý 3 - 2022', 'Quý 4 - 2022', 'Quý 1 - 2023',
    'Quý 2 - 2023', 'Quý 3 - 2023', 'Quý 4 - 2023', 'Quý 1 - 2024']

''' Duyệt qua lần lượt các hàng trong df, nếu giá trị của cột 'Quý' trong hàng đó bắt đầu bằng các chữ cái La Mã (I -> VI) 
thì kiểm tra xem giá trị của các cột trong list_quy nếu bằng 0 hoặc không có giá trị thì tính toán giá trị cho nó dựa trên công thức:
    giá trị ô đó = tổng các giá trị của các ô ở cùng cột nhưng trong các hàng phía dưới nó 
    có giá trị cột 'Quý' bắt đầu bằng số (1,2,...) cho đến khi gặp bắt đầu bởi chữ cái La Mã hoặc bắt đầu bởi chữ "T"
'''
for i in range(df.shape[0]):
    if df.loc[i, 'Quý'].startswith('I') or df.loc[i, 'Quý'].startswith('II') or df.loc[i, 'Quý'].startswith('III') or df.loc[i, 'Quý'].startswith('IV') or df.loc[i, 'Quý'].startswith('V') or df.loc[i, 'Quý'].startswith('VI'):
        for j in range(2, df.shape[1]):
            if df.loc[i, list_quy[j-2]] == 0:
                temp = i + 1
                while temp < df.shape[0] and not df.loc[temp, 'Quý'].startswith('I') and not df.loc[temp, 'Quý'].startswith('II') and not df.loc[temp, 'Quý'].startswith('III') and not df.loc[temp, 'Quý'].startswith('IV') and not df.loc[temp, 'Quý'].startswith('V') and not df.loc[temp, 'Quý'].startswith('VI') and not df.loc[temp, 'Quý'].startswith('T'):
                    df.loc[i, list_quy[j-2]] += df.loc[temp, list_quy[j-2]]
                    temp += 1

# Tính toán giá trị cho hàng 'A- TÀI SẢN NGẮN HẠN'
for quy in list_quy:
    df.loc[df['Quý'] == 'A- TÀI SẢN NGẮN HẠN', quy] = df.loc[df['Quý'] == 'I. Tiền và các khoản tương đương tiền', quy].sum() + df.loc[df['Quý'] == 'II. Các khoản đầu tư tài chính ngắn hạn', quy].sum() + df.loc[df['Quý'] == 'III. Các khoản phải thu ngắn hạn', quy].sum() + df.loc[df['Quý'] == 'IV. Hàng tồn kho', quy].sum() + df.loc[df['Quý'] == 'V. Tài sản ngắn hạn khác', quy].sum()

# Tính toán giá trị cho hàng 'B. TÀI SẢN DÀI HẠN'
for quy in list_quy:
    df.loc[df['Quý'] == 'B. TÀI SẢN DÀI HẠN', quy] = df.loc[df['Quý'] == 'I. Các khoản phải thu dài hạn', quy].sum() + df.loc[df['Quý'] == 'II.Tài sản cố định', quy].sum() + df.loc[df['Quý'] == 'III. Bất động sản đầu tư', quy].sum() + df.loc[df['Quý'] == 'IV. Tài sản dở dang dài hạn', quy].sum() + df.loc[df['Quý'] == 'V. Đầu tư tài chính dài hạn', quy].sum() + df.loc[df['Quý'] == 'VI. Tài sản dài hạn khác', quy].sum()

# Tính toán giá trị cho hàng 'C. NỢ PHẢI TRẢ'
for quy in list_quy:
    df.loc[df['Quý'] == 'C. NỢ PHẢI TRẢ', quy] = df.loc[df['Quý'] == 'I. Nợ ngắn hạn', quy].sum() + df.loc[df['Quý'] == 'II. Nợ dài hạn', quy].sum()

# Tính toán giá trị cho hàng 'D. VỐN CHỦ SỞ HỮU'
for quy in list_quy:
    df.loc[df['Quý'] == 'D. VỐN CHỦ SỞ HỮU', quy] = df.loc[df['Quý'] == 'I. Vốn chủ sở hữu', quy].sum() + df.loc[df['Quý'] == 'II. Nguồn kinh phí và quỹ khác', quy].sum()

# Tính toán giá trị cho hàng 'TỔNG CỘNG TÀI SẢN'
for quy in list_quy:
    df.loc[df['Quý'] == 'TỔNG CỘNG TÀI SẢN', quy] = df.loc[df['Quý'] == 'A- TÀI SẢN NGẮN HẠN', quy].sum() + df.loc[df['Quý'] == 'B. TÀI SẢN DÀI HẠN', quy].sum()

# Tính toán giá trị cho hàng 'TỔNG CỘNG NGUỒN VỐN'
for quy in list_quy:
    df.loc[df['Quý'] == 'TỔNG CỘNG NGUỒN VỐN', quy] = df.loc[df['Quý'] == 'C. NỢ PHẢI TRẢ', quy].sum() + df.loc[df['Quý'] == 'D. VỐN CHỦ SỞ HỮU', quy].sum()

# Lưu các giá trị đã tính toán vào file excel
df.to_excel('bctc/Can_doi_ke_toan_VCB.xlsx', index=False)
```

```python
import pandas as pd
def check_full_data(excel_path):
    df = pd.read_excel(excel_path, header=1)

    list_quy = ['Quý 1 - 2021',
    'Quý 2 - 2021', 'Quý 3 - 2021', 'Quý 4 - 2021', 'Quý 1 - 2022',
    'Quý 2 - 2022', 'Quý 3 - 2022', 'Quý 4 - 2022', 'Quý 1 - 2023',
    'Quý 2 - 2023', 'Quý 3 - 2023', 'Quý 4 - 2023', 'Quý 1 - 2024']

    # Tính toán giá trị cho hàng 'A- TÀI SẢN NGẮN HẠN'
    for quy in list_quy:
        # Lấy giá trị của 'TỔNG CỘNG TÀI SẢN' và 'TỔNG CỘNG NGUỒN VỐN'
        total_tai_san = df.loc[df['Quý'] == 'TỔNG CỘNG TÀI SẢN', quy].values
        total_nguon_von = df.loc[df['Quý'] == 'TỔNG CỘNG NGUỒN VỐN', quy].values
        
        # Kiểm tra nếu gái trị 'TỔNG CỘNG TÀI SẢN' hoặc 'TỔNG CỘNG NGUỒN VỐN' rỗng hoặc không bằng nhau thì trả về False
        if len(total_tai_san) == 0 or len(total_nguon_von) == 0 or total_tai_san[0] != total_nguon_von[0]:
            return False
    return True

list_company_full = []

# Duyệt qua từng file excel trong thư mục bctc, nếu file đó đã được tính toán đầy đủ thì thêm tên file vào list_company_full
import os
for file in os.listdir('bctc'):
    if file.endswith('.xlsx'):
        if check_full_data(f'bctc/{file}'):
            list_company_full.append(file)

print(list_company_full)

# Lấy ra danh sách mã cổ phiếu của các công ty đã tính toán đầy đủ dữ liệu
list_ma_cp_full = []
for file in list_company_full:
    # Lấy ra mã cổ phiếu từ tên file (ba ký tự cuối cùng)
    ma_cp = file.split('.')[0][-3:]
    list_ma_cp_full.append(ma_cp)

print(list_ma_cp_full)
print(len(list_ma_cp_full))
```

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Dữ liệu từ Quý 2 - 2018 đến Quý 1 - 2024
data = [3372708212260, 3777607549818, 4576157128799, 4786813479601, 
        6489699296640, 6685445083032, 7529166722403,7982072749097, 
        8351307712921, 7938063247638, 7990130386206, 7768231726599, 
        7786584839414, 8206176494747, 8569613917893, 9179385142444, 
        9642310081327, 9957115220854, 9991631928569, 10477077016930, 
        10928357909053, 11277995454318, 10781316431919, 10901971260367, 
        11231809771863, 11610261843767, 11534976245894, 11686618010689
]

# Tạo DataFrame với cột 'ds' (ngày tháng) và 'y' (giá trị)
dates = pd.date_range(start='2017-04-01', periods=len(data), freq='Q')
df = pd.DataFrame({'ds': dates, 'y': data})

# Khởi tạo mô hình Prophet
model = Prophet(seasonality_mode='multiplicative', yearly_seasonality=True)

# Huấn luyện mô hình
model.fit(df)

# Dự báo cho 1 bước quý tiếp theo (Quý 2 - 2024)
future = model.make_future_dataframe(periods=1, freq='Q')
forecast = model.predict(future)

# Lấy giá trị dự báo cho Quý 2 - 2024
predicted_value = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(1)
print("Dự báo cho Quý 2 - 2024:")
print(predicted_value)

# Vẽ biểu đồ dữ liệu gốc và dự báo
fig1 = model.plot(forecast)
plt.xlabel("Thời gian")
plt.ylabel("Giá trị")
plt.title("Dự báo giá trị cho Quý 2 - 2024")

# Vẽ các thành phần xu hướng và mùa vụ
fig2 = model.plot_components(forecast)
plt.show()
```
